```python
!pip install -q aitextgen #install the main package
from aitextgen import aitextgen
import pandas as pd
ai = aitextgen(tf_gpt2="124M", to_gpu=True)

df = pd.read_csv('../input/eluvio-dataset/Eluvio_DS_Challenge.csv')
df = df.sort_values(by='up_votes').reset_index(drop=True)
df = df[200000:]

df.to_csv("input_text_cleaned.txt", columns=["title"],sep='\n' ,header=False, index=False)
ai.train('input_text_cleaned.txt',
         line_by_line=False,
         num_steps=10000,
         generate_every=100,
         save_every=500,
         learning_rate=1e-3,
         batch_size=1, 
         )

ai = aitextgen(model="./trained_model/pytorch_model.bin", config="./trained_model/config.json", to_gpu=True)
ai.generate(n=3,prompt="Russia",
            batch_size=1,
            max_length=50,
            temperature=1.0,
            top_p=0.9)
```
